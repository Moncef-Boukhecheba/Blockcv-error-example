### Loading the dataset

```r
# Load species data
species_data_final <- readRDS('test_blockcv.rds')
```

These data are in the **SpatialPointsDataFrame** format.

They contain two components: 
- Coords, which contains the latitude longitude pairs
- and Data, which contains here the data on the climatic variables

They can be displayed :

```r
summary(species_data_final)
```

```r
species_data_final@coords
```

```r
species_data_final@data
```

```r
folds <- spatialBlock(speciesData = species_data_final, 
                       species = "Present", 
                       theRange = 500, 
                       k = 5,
                       showBlocks = TRUE, 
                       verbose = TRUE,
                       selection = "random", 
                       iteration = 100, 
                       biomod2Format = TRUE)
```
